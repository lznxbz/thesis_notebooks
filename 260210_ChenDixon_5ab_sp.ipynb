```python
import pandas as pd, re
from pathlib import Path

from parse_dftsp_grepscf_to_df import parse_grepscf

hartree_to_eV = 27.2114
```

# Compared DFT sp energies
pbe0/def2tzvp pbe/def2tzvp

```python
energy_file_path = Path('./tio2_ChenDixon_gjf/all5_sp/sp_energy.txt')
df_sp_energies = parse_grepscf(energy_file_path)
```

```python
df_sp_energies
```

```python
df_5apbe0 = df_sp_energies[df_sp_energies["structure"] == "5a"].loc[df_sp_energies["sp method"] == "pbe0_def2tzvp"]
df_5bpbe0 = df_sp_energies[df_sp_energies["structure"] == "5b"].loc[df_sp_energies["sp method"] == "pbe0_def2tzvp"]
df_5apbe = df_sp_energies[df_sp_energies["structure"] == "5a"].loc[df_sp_energies["sp method"] == "pbe_def2tzvp"]
df_5bpbe = df_sp_energies[df_sp_energies["structure"] == "5b"].loc[df_sp_energies["sp method"] == "pbe_def2tzvp"]
```

```python
for df_i in [df_5apbe0, df_5bpbe0,df_5apbe, df_5bpbe]:
    df_i.insert(0, "opt method", df_i.pop("opt method"))
    df_i.insert(0, "sp method", df_i.pop("sp method"))
    df_i.reset_index(drop=True, inplace=True)
```

```python
df_5apbe0
```

```python
df_5bpbe0
```

```python
df_5bpbe0_x = df_5bpbe0.iloc[:,2:]
df_pbe0 = pd.concat([df_5apbe0, df_5bpbe0_x], axis=1)
df_pbe0.columns.values[3] = '5a_energy_au'
df_pbe0.columns.values[5] = '5b_energy_au'
df_pbe0.drop('structure', inplace=True, axis=1)
df_pbe0
```

```python
df_5bpbe_x = df_5bpbe.iloc[:,2:]
df_pbe = pd.concat([df_5apbe, df_5bpbe_x], axis=1)
df_pbe.columns.values[3] = '5a_energy_au'
df_pbe.columns.values[5] = '5b_energy_au'
df_pbe.drop('structure', inplace=True, axis=1)
df_pbe
```

```python
for df_i in [df_pbe0, df_pbe]:
    df_i.loc[:, "dE_eV"] = (df_i.loc[:, "5b_energy_au"] - df_i.loc[:, "5a_energy_au"]) * hartree_to_eV
    print(df_i)
    # e_ref = df_i[df_i["opt method"] == "pbe0_def2tzvp", "5a_energy_au"]
    # df_i.loc[:, "5a_energy_au_ref"]
```

```python
e_ref_a = df_pbe0.loc[df_pbe0["opt method"] == "pbe0_def2tzvp", "5a_energy_au"].iloc[0]
e_ref_b = df_pbe0.loc[df_pbe0["opt method"] == "pbe0_def2tzvp", "5b_energy_au"].iloc[0]
print(e_ref_a)
print(e_ref_b)

df_pbe0.loc[:,"5a_energy_eV_ref"] = (df_pbe0.loc[:,"5a_energy_au"] - e_ref_a) * hartree_to_eV
df_pbe0.loc[:,"5b_energy_eV_ref"] = (df_pbe0.loc[:,"5b_energy_au"] - e_ref_b) * hartree_to_eV

df_pbe0.loc[:,"dE_eV_ref"] = df_pbe0.loc[:,"5b_energy_eV_ref"] - df_pbe0.loc[:,"5a_energy_eV_ref"]
df_pbe0
```

```python
e_ref_a = df_pbe.loc[df_pbe["opt method"] == "pbe_def2tzvp", "5a_energy_au"].iloc[0]
e_ref_b = df_pbe.loc[df_pbe["opt method"] == "pbe_def2tzvp", "5b_energy_au"].iloc[0]
print(e_ref_a)
print(e_ref_b)

df_pbe.loc[:,"5a_energy_eV_ref"] = (df_pbe.loc[:,"5a_energy_au"] - e_ref_a) * hartree_to_eV
df_pbe.loc[:,"5b_energy_eV_ref"] = (df_pbe.loc[:,"5b_energy_au"] - e_ref_b) * hartree_to_eV

df_pbe.loc[:,"dE_eV_ref"] = df_pbe.loc[:,"5b_energy_eV_ref"] - df_pbe.loc[:,"5a_energy_eV_ref"]
df_pbe
```
